```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)        # Don't wrap lines
pd.set_option('display.max_colwidth', None) # Show full content in each column
```

```python
cards_data=pd.read_csv("cards_data.csv")
cards_data.head()
```

```python
evaluation_features=pd.read_csv("evaluation_features.csv")
evaluation_features.head()
```

```python
transactions_train=pd.read_csv("transactions_train.csv")
transactions_train.head()
```

```python
users_data=pd.read_csv("users_data.csv")
users_data.head()
```

```python
m1_train = pd.merge(
    transactions_train,
    users_data,
    how='left',             # type de jointure
    left_on='client_id',    # clé de jointure dans transaction_data
    right_on='id'           # clé de jointure dans client_data
)

m2_train = pd.merge(
    m1_train,
    cards_data,
    how='left',             
    left_on='card_id',    
    right_on='id'           
)

X_train=m2_train
```

```python
y_dic = pd.read_json("train_fraud_labels.json")
y_m1 = y_dic.loc[X_train['transaction_id'], 'target']

y_m1.to_pickle("y_train.pkl")
```

```python
X_train.columns
```

```python
X_train['date'] = pd.to_datetime(X_train['date'])
X_train['year'] = X_train['date'].dt.year
X_train['month'] = X_train['date'].dt.month
X_train['day'] = X_train['date'].dt.day
X_train['hour'] = X_train['date'].dt.hour
X_train['minute'] = X_train['date'].dt.minute
X_train['second'] = X_train['date'].dt.second

X_train['expires'] = pd.to_datetime(X_train['date'])
X_train['expires month'] = X_train['expires'].dt.month
X_train['expires year'] = X_train['expires'].dt.year

X_train['acct_open_date'] = pd.to_datetime(X_train['date'])
X_train['acct_open_date year'] = X_train['acct_open_date'].dt.year
X_train['acct_open_date month'] = X_train['acct_open_date'].dt.month
X_train['acct_open_date day'] = X_train['acct_open_date'].dt.day
X_train['acct_open_date hour'] = X_train['acct_open_date'].dt.hour
X_train['acct_open_date minute'] = X_train['acct_open_date'].dt.minute

X_train['amount'] = X_train['amount'].str.replace('$', '', regex=False).astype(float)
X_train['per_capita_income'] = X_train['per_capita_income'].str.replace('$', '', regex=False).astype(float)
X_train['total_debt'] = X_train['total_debt'].str.replace('$', '', regex=False).astype(float)
X_train['yearly_income'] = X_train['yearly_income'].str.replace('$', '', regex=False).astype(float)
X_train['credit_limit'] = X_train['credit_limit'].str.replace('$', '', regex=False).astype(float)

X_train = X_train.drop(columns=['transaction_id', 'address', 'date', 'expires', 'acct_open_date'])
```
